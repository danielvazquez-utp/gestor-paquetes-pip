# Gestor de paquetes de Python - PIP

PIP (Prefered Installation Program) es un gestor para la instalación de paquetes en el entorno de desarrollo de Python. Cada entorno de desarrollo puede contener uno o mas paquetes o módulos. Si bien, la instalación base de Python contiene algunos paquetes por defecto, para la implementación de algunos objetos especiales como los *DataFrames*, *Matrices*, *Vectores*, entre otros, requieren una instalación adicional.

*PIP* incluye algunas directivas de uso frecuente los cuales ayudan a realizar algunas tareas específicas. Para ver un listado general de estas directivas se puede utilizar *pip --help*.

```python
# Verificación de la versión de PIP
!pip --version
```

La instalación de paquetes mediante *pip* tiene la siguiente sintaxis: 

```python
pip install <paquete 1|módulo 1> ... <paquete n|módulo n> 
```

```python
!pip install numpy pandas
```

```python
# Listado de paquetes
!pip list
```

```python
# Listado de paqueterpias y módulos cob freeze
!pip freeze
```

```python
# Creación de un archivo de requisitos
!pip freeze > requeriments.txt
```
