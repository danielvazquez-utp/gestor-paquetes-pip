# Pandas

*Pandas* es un paquete especializado en la manipulación y gestión de datos a través de dos estructuras: *Serie* y *DataFrame*.

```python
# Instalación de pandas
pip install pandas
```

```python
# Importación
import pandas as pd
```

*Pandas* emplea algunas estructuras del paquete *numpy* como los *ndarrays* para la creación de sus propias estructuras de datos.

## Series

Una *serie* es un *ndarray* o colección lineal la cual tienen asociados indices con valores.

```python
# Definición de una serie
pd.Series([1, 2, 3], name='integer')
```

### Atributos

Las *series* contiene una gran cantidad de atributos de los cuales destacan los siguientes:

```python
estaturas = pd.Series({
    "Daniel": 1.75,
    "Oscar": 1.79,
    "Ana": 1.54,
    "Javier": 1.84,
    "Fernanda": 1.67
}, name="Estaturas de los estudiantes")
print(estaturas)
```

```python
# Recuperar la lista de indices de una serie
estaturas.index
```

Para recuperar elementos contenidos en una *serie* se pueden utilizar diferentes métodos, por ejemplo, tomando en consideración la posición del elemento utilizando el atributo *.iloc[]*

```python
# Selección de elementos con iloc[]
estaturas.iloc[0]
```

```python
estaturas.iloc[-1]
```

```python
estaturas.iloc[1]
```

```python
estaturas.iloc[0:3]
```

```python
estaturas.iloc[2:]
```

Sin embargo, si se desea recuperar un elemento mediante su clave se utiliza el atributo *.loc[]*

```python
# Recuperando elementos con .loc[]
estaturas.loc["Ana"]
```

```python
# Recuperando un rango de valores con .loc[]
estaturas.loc["Daniel": "Javier"]
```

Para obtener una muestra de una *serie* se pueden emplear dos atributos: *head()* y *tail()*. 

```python
# Recuperando una muestra de los primeros elementos de la serie datos con head()
estaturas.head(3)
```

```python
# Recuperando una muestra de datos de los ultimos elementos de la serie con tail()
estaturas.tail(2)
```
